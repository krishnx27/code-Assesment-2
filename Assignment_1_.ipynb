```python

import numpy as np
import matplotlib.pyplot as plt
import numpy as np
import skimage as io
import cv2
from google.colab.patches import cv2_imshow
```

```python

image_1 = cv2.imread("/content/cracked sand.jpg")
image_1 = np.array(image_1)
gray_1 = cv2.cvtColor(image_1,cv2.COLOR_BGR2GRAY)

image_2 = cv2.imread("/content/white sand.jpg")
image_2 = np.array(image_2)
gray_2 = cv2.cvtColor(image_2,cv2.COLOR_BGR2GRAY)
```

```python

fig, ax = plt.subplots(1, 2, figsize=(10, 5))
fig.tight_layout()
ax[0].imshow(cv2.cvtColor(image_1, cv2.COLOR_BGR2RGB))
ax[0].set_title('Image 1')

ax[1].imshow(gray_1, cmap='gray')  # Set colormap to grayscale
ax[1].set_title('Grayscale Image')
plt.show()
```

```python

fig, ax = plt.subplots(1, 2, figsize=(10, 5))
fig.tight_layout()
ax[0].imshow(cv2.cvtColor(image_2, cv2.COLOR_BGR2RGB))
ax[0].set_title('Image 2')

ax[1].imshow(gray_2, cmap='gray')  # Set colormap to grayscale
ax[1].set_title('Grayscale Image')
plt.show()
```

```python

def calculate_lbp(img, x, y):
    center = img[x][y]
    values = []
    for i in range(-1, 2):
        for j in range(-1, 2):
            if i == 0 and j == 0:
                continue
            values.append(1 if img[x + i][y + j] >= center else 0)
    val = 0
    for i in range(len(values)):
      val+=values[i]*2**i

    return val
```

```python
def apply_lbp(img):
    height, width = img.shape
    lbp_img = np.zeros((height, width), dtype=np.uint8)
    for i in range(1, height - 1):
        for j in range(1, width - 1):
            lbp_img[i, j] = calculate_lbp(img, i, j)
    return lbp_img
```

```python
lbp_image_1 = apply_lbp(gray_1)
lbp_image_2 = apply_lbp(gray_2)
```

```python

fig, ax = plt.subplots(1, 2, figsize=(10, 5))
fig.tight_layout()
ax[0].imshow(lbp_image_1, cmap='gray')
ax[0].set_title('lbp_image_1')
ax[1].imshow(lbp_image_2, cmap='gray')
ax[1].set_title('lbp_image_2')
plt.show()
```
